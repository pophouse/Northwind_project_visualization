ORDER FULFILMENT RATE

```python
%%sql

WITH SupplierPerformance AS (
    SELECT 
        s.SupplierID,
        s.CompanyName AS SupplierName,
        COUNT(DISTINCT o.OrderID) AS TotalOrders,
        COUNT(DISTINCT CASE 
            WHEN o.ShippedDate <= o.RequiredDate THEN o.OrderID 
        END) AS OnTimeOrders
    FROM Suppliers s
    JOIN Products p ON s.SupplierID = p.SupplierID
    JOIN [OrderDetails] od ON p.ProductID = od.ProductID
    JOIN Orders o ON od.OrderID = o.OrderID
    GROUP BY s.SupplierID, s.CompanyName
)
SELECT 
    SupplierID,
    SupplierName,
    TotalOrders,
    OnTimeOrders,
    ROUND((OnTimeOrders * 100.0) / NULLIF(TotalOrders, 0), 2) AS FulfillmentRate_Percentage
FROM SupplierPerformance
ORDER BY FulfillmentRate_Percentage DESC
LIMIT 10
```

## PROFITABILITY ANALYSIS

```python
%%sql

SELECT 
    c.CategoryName,
    p.ProductName,
    SUM (ROUND(od.UnitPrice * od.Quantity))  AS Gross_sale,
    COUNT(od.OrderID) AS profit
FROM 
    OrderDetails od 
JOIN
    Products p ON od.ProductID = p.ProductID
     
JOIN
    Categories c ON p.CategoryID = c.CategoryID 

 GROUP BY
    c.CategoryName, p.ProductName
ORDER BY 
    Gross_sale DESC 
limit 10
```

```python
%%sql

SELECT
    p.ProductName,
    c.CategoryName AS Product_Category,
    SUM(od.UnitPrice * od.Quantity * (1 - od.Discount)) AS Sales_Revenue,
    SUM(p.UnitPrice * od.Quantity) AS Total_Sales,
    SUM((od.UnitPrice * od.Quantity * (1 - od.Discount)) - (p.UnitPrice * od.Quantity)) AS Gross_Profit,
    ROUND(SUM((od.UnitPrice * od.Quantity * (1 - od.Discount)) - (p.UnitPrice * od.Quantity)) / 
          NULLIF(SUM(od.UnitPrice * od.Quantity * (1 - od.Discount)), 0) * 100, 2) AS Gross_Profit_Margin_Percentage
FROM 
    OrderDetails od
JOIN 
    Products p ON od.ProductID = p.ProductID
JOIN
    Categories c ON p.CategoryID = c.CategoryID
GROUP BY p.ProductName, c.CategoryName
ORDER BY Gross_Profit DESC
LIMIT 10;
```

```python
import sqlite3
import pandas as pd
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
from statsmodels.tsa.arima.model import ARIMA
import plotly.express as px
conn = sqlite3.connect("Northwind.db")
```

```python
orders =  pd.read_sql("SELECT * FROM Orders", conn)             
order_details =  pd.read_sql("SELECT * FROM OrderDetails", conn)
products =  pd.read_sql("SELECT * FROM Products", conn)
categories = pd.read_sql('SELECT * FROM Categories', conn)
sales = orders.merge(order_details, on='OrderID', how='inner') \
              .merge(products[['ProductID', 'ProductName']], on='ProductID',how='left')
```

```python
print(sales.isnull().sum())
```

```python
sales['ShippedDate'] = sales['ShippedDate'].fillna(0)
sales['ShipRegion'] = sales['ShipRegion'].fillna('Unknown Region')
sales['ShipPostalCode'] = sales['ShipPostalCode'].fillna(00000)
```

```python
print(sales.isnull().sum())
```

## Aggregate sales into monthly

```python
# Merge data into a comprehensive DataFrame
df = pd.merge(orders, order_details, on='OrderID')
df = pd.merge(df, products, on='ProductID')
df = pd.merge(df, categories, on='CategoryID')

# Convert dates and calculate revenue
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['Sales'] = df['Quantity'] * df['UnitPrice_x']
```

```python
monthly_sales = df.groupby(pd.Grouper(key='OrderDate', freq='ME'))['Sales'].sum().reset_index()

monthly_sales.columns = ['Month', 'TotalSales']

monthly_sales.set_index('Month', inplace=True)

print(monthly_sales.head())
```

## Check for Seasonality and Trend

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
monthly_sales.plot()
plt.title('Monthly Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(monthly_sales, model='additive',period =11)
decomposition.plot()
plt.show()
```

```python
from pandas.plotting import autocorrelation_plot

autocorrelation_plot(monthly_sales)
plt.title("Autocorrelation Plot")
plt.show()
```

```python
#check for trends with Adf test
from statsmodels.tsa.stattools import adfuller

result = adfuller(monthly_sales)
print("ADF Statistic:", result[0])
print("p-value:", result[1])
if result[1] < 0.05:
    print("Series is stationary → No trend")
else:
    print("Series is non-stationary → Trend likely present")
```

### Checking for Anomalies using z-score

```python
#detect Anomalies
decomp = decomposition
resid = decomp.resid.dropna()
threshold = 2 * resid.std()
anomalies = resid[np.abs(resid) > threshold]

# Plot anomalies
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales, label='Original')
plt.scatter(anomalies.index, anomalies, color='red', label='Anomalies')
plt.legend()
plt.title("Monthly Sales With Anomalies")
plt.show()
```

## TRAIN AND VALIDATE MOVING AVERAGE MODEL

```python
np.random.seed(0)
dates = pd.date_range(start ='1996-01-01' , periods=24, freq='ME')
values = 100 + np.random.randn(24).cumsum()
df = pd.DataFrame({'ds': dates, 'y': values})
```

```python
train = df.iloc[:-12].copy()
valid = df.iloc[-12:].copy()
```

```python
plt.plot(train['ds'], train['y'], label ='Train')
plt.plot(valid['ds'], valid['y'], label ='Actual')
plt.plot(valid['ds'], valid['yhat'],'--', label ='SIMPLE MOVING AVERAGE')
plt.legend(), 
plt.title('SMA forecast')
plt.tight_layout()
plt.show()
```

## model evaluation for Moving Average

```python
from sklearn.metrics import mean_absolute_error, root_mean_squared_error

mae = mean_absolute_error(valid['y'], valid['yhat'])
rmse = root_mean_squared_error(valid['y'], valid['yhat'])
print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, Holt, ExponentialSmoothing
# Sample time series: monthly data
np.random.seed(0)
data = pd.Series(
    50 + np.arange(36) * 0.5 + 5 * np.sin(np.linspace(0, 3 * np.pi, 36)) + np.random.normal(scale=2, size=36),
    index=pd.date_range(start='2020-01-01', periods=36, freq='ME')
)

# --- 1. Simple Exponential Smoothing ---
ses_model = SimpleExpSmoothing(data).fit(smoothing_level=0.2, optimized=False)
ses_forecast = ses_model.forecast(12)

# --- 2. Holt’s Linear Trend Method ---
holt_model = Holt(data).fit(smoothing_level=0.8, smoothing_trend=0.2, optimized=False)
holt_forecast = holt_model.forecast(12)

# --- 3. Holt-Winters Seasonal Method (Additive) ---
hw_model = ExponentialSmoothing(
    data,
    trend='add',
    seasonal='add',
    seasonal_periods=12
).fit()
hw_forecast = hw_model.forecast(12)

# --- Plotting ---
plt.figure(figsize=(12, 6))
plt.plot(data, label='Original Data', color='black')
plt.plot(ses_forecast, label='SES Forecast', linestyle='--')
plt.plot(holt_forecast, label='Holt Forecast', linestyle='--')
plt.plot(hw_forecast, label='Holt-Winters Forecast', linestyle='--')
plt.title('Exponential Smoothing Forecasts')
plt.legend()
plt.grid(True)
plt.show()
```

## training and validation of Exponential smoothing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# 1. Generate sample time series data
np.random.seed(42)
dates = pd.date_range(start='2020-01-01', periods=48, freq='ME')
series = pd.Series(100 + np.random.randn(48).cumsum(), index=dates)

# 2. Split into train and validation sets
train = series.iloc[:36]
valid = series.iloc[36:].copy()  # Make a copy to avoid SettingWithCopyWarning

# 3. Fit Exponential Smoothing model
model = ExponentialSmoothing(train, trend='add', seasonal=None)
fit = model.fit()

# 4. Forecast and insert into validation using .iloc
forecast = fit.forecast(steps=len(valid))
valid_forecast = valid.copy()
valid_forecast.iloc[:] = forecast.values  # Assign forecast using .iloc

# 5. Plot
plt.plot(train.index, train, label='Train')
plt.plot(valid.index, valid, label='Actual')
plt.plot(valid_forecast.index, valid_forecast, '--', label='Forecast')
plt.title('Exponential Smoothing: Forecast')
plt.legend()
plt.tight_layout()
plt.show()
```

## model evaluation for ES

```python
mae = mean_absolute_error(valid, forecast)
rmse = root_mean_squared_error(valid, forecast)

print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
```
